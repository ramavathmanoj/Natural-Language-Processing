```python
#Import necessary libraries and download VADER lexicon
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon needed for sentiment analysis
nltk.download('vader_lexicon')
```

```python
#Define the function for sentiment scoring
def sentiment_scores(sentence):
    # Create a SentimentIntensityAnalyzer object
    sid_obj = SentimentIntensityAnalyzer()

    # Get polarity scores for the input sentence
    sentiment_dict = sid_obj.polarity_scores(sentence)

    # Print the full sentiment dictionary
    print("Sentence sentiment analysis:", sentiment_dict)

    # Print the percentage of negative sentiment
    print(sentiment_dict['neg'] * 100, "% Negative")

    # Print the percentage of neutral sentiment
    print(sentiment_dict['neu'] * 100, "% Neutral")

    # Print the percentage of positive sentiment
    print(sentiment_dict['pos'] * 100, "% Positive")

    # Determine overall sentiment based on compound score
    print("Sentence Overall Rated As", end=" ")
    if sentiment_dict['compound'] >= 0.05:
        print("Positive")
    elif sentiment_dict['compound'] <= -0.05:
        print("Negative")
    else:
        print("Neutral")
```

```python
#Import additional libraries and download VADER lexicon
import pandas as pd
import re
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon (again to ensure it's available)
nltk.download('vader_lexicon')
```

```python
#Define function with sentiment result returned
def sentiment_scores(sentence):
    # Create SentimentIntensityAnalyzer object
    sid_obj = SentimentIntensityAnalyzer()

    # Get polarity scores
    sentiment_dict = sid_obj.polarity_scores(sentence)

    # Print the sentiment dictionary
    print("Sentence sentiment analysis:", sentiment_dict)

    # Print each sentiment percentage
    print(sentiment_dict['neg'] * 100, "% Negative")
    print(sentiment_dict['neu'] * 100, "% Neutral")
    print(sentiment_dict['pos'] * 100, "% Positive")

    # Determine overall sentiment
    print("Sentence Overall Rated As", end=" ")
    if sentiment_dict['compound'] >= 0.05:
        print("Positive")
        label = "Positive"
    elif sentiment_dict['compound'] <= -0.05:
        print("Negative")
        label = "Negative"
    else:
        print("Neutral")
        label = "Neutral"

    # Return both label and compound score
    return label, sentiment_dict['compound']
```

```python
# Load the CSV file containing tweets
df = pd.read_csv("tweets-data.csv")

# Randomly select 500 tweets for analysis
df_sample = df.sample(n=500, random_state=42).reset_index(drop=True)
```

```python
#Clean the tweets text
def clean_text(text):
    text = str(text)

    # Remove URLs
    text = re.sub(r"http\\S+", "", text)

    # Remove Twitter mentions
    text = re.sub(r"@\\w+", "", text)

    # Remove hashtags symbol but keep the words
    text = re.sub(r"#", "", text)

    # Remove any character that is not a letter or whitespace
    text = re.sub(r"[^A-Za-z\\s]", "", text)

    # Convert text to lowercase and remove extra spaces
    text = text.lower().strip()

    return text

# Apply text cleaning to tweets
df_sample['cleaned_text'] = df_sample['Tweets'].apply(clean_text)
```

```python
#Apply sentiment analysis to cleaned tweets
# Apply sentiment analysis function to cleaned text
# Store both sentiment label and sentiment score in new columns
df_sample[['sentiment_label', 'sentiment_score']] = df_sample['cleaned_text'].apply(lambda x: pd.Series(sentiment_scores(x)))
```
